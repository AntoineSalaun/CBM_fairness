## Common functions definition

```python
def read_accuracy_from_txt(path_document):
    with open(path_document, 'r') as file:
        line = file.readline().strip()
        return float(line)
    
def read_val_accuracy_baseline(path_model):
    with open(path_model, 'r') as file:
        line = file.readline().strip()
        try:
            # Split the line by the colon and take the second part
            accuracy = float(line.split(":")[1].strip())
            return accuracy
        except (IndexError, ValueError):
            return None
        
def read_val_accuracy_cbm(path_model):
    with open(path_model, 'r') as file:
        data = json.load(file)
        acc_val = data["metrics"]["acc_val"]
        return acc_val
```

# Values on val

## 30 verbs baseline

```python
path_baseline = Path.cwd() / 'saved_models/imSitu/30_verbs/grid_search/baseline'
```

```python
lr=[0.001, 0.01, 0.1]
momentum=[0.8, 0.9, 0.95]
step_size=[5, 7, 10]
gamma=[0.1, 0.5, 0.9]

best_accuracy_balanced = -1
best_accuracy_imbalanced_1 = -1 
best_accuracy_imbalanced_2 = -1
best_accuracy_aggregate = -1

for lr_ in lr:
    for momentum_ in momentum: 
        for step_size_ in step_size:
            for gamma_ in gamma: 
                name_folder = f'lr_{lr_}_momentum_{momentum_}_step_{step_size_}_gamma_{gamma_}'
                path_folder = path_baseline / name_folder
                
                path_balanced = path_folder / 'balanced'
                path_imbalanced_1 = path_folder / 'imbalanced_1'
                path_imbalanced_2 = path_folder / 'imbalanced_2'
                
                accuracy_balanced = read_val_accuracy_baseline(path_balanced / 'results.txt')
                accuracy_imbalanced_1 = read_val_accuracy_baseline(path_imbalanced_1 / 'results.txt')
                accuracy_imbalanced_2 = read_val_accuracy_baseline(path_imbalanced_2 / 'results.txt')
                accuracy_aggregate = (accuracy_balanced + accuracy_imbalanced_1 + accuracy_imbalanced_2) / 3
                
                if accuracy_balanced > best_accuracy_balanced:
                    best_accuracy_balanced = accuracy_balanced
                    best_accuracy_b_hyperparameters = [lr_, momentum_, step_size_, gamma_]
                    
                if accuracy_imbalanced_1 > best_accuracy_imbalanced_1:
                    best_accuracy_imbalanced_1 = accuracy_imbalanced_1
                    best_accuracy_im1_hyperparameters = [lr_, momentum_, step_size_, gamma_]
                
                if accuracy_imbalanced_2 > best_accuracy_imbalanced_2:
                    best_accuracy_imbalanced_2 = accuracy_imbalanced_2
                    best_accuracy_im2_hyperparameters = [lr_, momentum_, step_size_, gamma_]
                    
                if accuracy_aggregate > best_accuracy_aggregate:
                    best_accuracy_aggregate = accuracy_aggregate
                    best_accuracy_agg_hyperparameters = [lr_, momentum_, step_size_, gamma_]
                    
                    
print(best_accuracy_balanced)
print(best_accuracy_imbalanced_1)
print(best_accuracy_imbalanced_2)
print(best_accuracy_aggregate)

print('______________________')

print(best_accuracy_b_hyperparameters)
print(best_accuracy_im1_hyperparameters)
print(best_accuracy_im2_hyperparameters)
print(best_accuracy_agg_hyperparameters)
```

```python
def read_results(path_result, print_=True):

    path_balanced = path_result / 'balanced'
    path_imbalanced_1 = path_result / 'imbalanced_1'
    path_imbalanced_2 = path_result / 'imbalanced_2'

    accuracy_balanced = read_accuracy_from_txt(path_balanced / 'accuracy.txt')
    accuracy_imbalanced_1 = read_accuracy_from_txt(path_imbalanced_1 / 'accuracy.txt')
    accuracy_imbalanced_2 = read_accuracy_from_txt(path_imbalanced_2 / 'accuracy.txt')
    accuracy_aggregate = (accuracy_balanced + accuracy_imbalanced_1 + accuracy_imbalanced_2) / 3
    
    if print_:
        print(f'Balanced: {accuracy_balanced}, Imbalanced 1: {accuracy_imbalanced_1}, Imbalanced 2: {accuracy_imbalanced_2}, Aggregate: {accuracy_aggregate}')

    return accuracy_balanced, accuracy_imbalanced_1, accuracy_imbalanced_2, accuracy_aggregate

def read_results_full(path_result, print_=True):
    
    accuracy = read_accuracy_from_txt(path_result / 'accuracy.txt')
    if print_:
        print(f'Accuracy: {accuracy}')
        
    return accuracy

def read_result_csv(path_result):
    df =  pd.read_csv(path_result / 'accuracies_by_group.csv')
    df['Accuracy Parity	'] = df['Accuracy Parity'] * 100
    return df

def read_3_results(path_result):
#    df_balanced = read_result_csv(path_result / 'balanced' / 'accuracies_by_group.csv')
 #   df_imbalanced_1 = read_result_csv(path_result / 'imbalanced_1' / 'accuracies_by_group.csv')
  #  df_imbalanced_2 = read_result_csv(path_result / 'imbalanced_2' / 'accuracies_by_group.csv')

    
    df_balanced = pd.read_csv(path_result / 'balanced' / 'accuracies_by_group.csv')
    df_imbalanced_1 = pd.read_csv(path_result / 'imbalanced_1' / 'accuracies_by_group.csv')
    df_imbalanced_2 = pd.read_csv(path_result / 'imbalanced_2' / 'accuracies_by_group.csv')

    return df_balanced, df_imbalanced_1, df_imbalanced_2
```
